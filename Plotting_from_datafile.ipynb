# Plotting total DOSS

```python
import matplotlib
import itertools
import matplotlib.pyplot as plt
import numpy as np

X, Y = [], []

# Reading data from .f25 file (DOSS)
with open('Si/Si_dat.DOSS', 'r') as f:
    for line in itertools.islice(f, 5, 202): # Start reading at line 5 and stop at line 202
      values = [float(s) for s in line.split()]
      X.append(values[0]*27.21) # Energy in Ha is converted into eV
      Y.append(values[1])
    xlow=min(X)-0.5
    xhigh=max(X)+0.5
    ylow=0.
    yhigh=max(Y)+1

# Plotting data
fig=plt.figure()
ax=plt.subplot()
ax.tick_params(direction='in',which='both')
ax.plot(X,Y)
ax.set(xlabel='$E-E_{fermi}$ (eV)', ylabel='$n(E-E_{fermi})$ (arb. unit)',
       title='Total density of states')
plt.axvline(x=0, color='r', linestyle='-') # Vertical red line in x = 0
plt.xlim(xlow,xhigh) # Set the x range
plt.ylim(ylow,yhigh) # Set the y range
plt.show()
fig.savefig('Si/Si_DOSS.png',dpi=600)
```

# Plotting DOSS projected on each atom (only for MgO)

```python
X, Y1, Y2, Y3 = [], [], [], []

# Reading data from .f25 file (DOSS)
with open('MgO/MgO_dat.DOSS', 'r') as f:
    for line in itertools.islice(f, 5, 202): # Start reading at line 5 and stop at line 202 ()
      values = [float(s) for s in line.split()]
      X.append(values[0]*27.21)
      Y1.append(values[1])
      Y2.append(values[2])
      Y3.append(values[3])
    xlow=min(X)-0.5
    xhigh=max(X)+0.5
    ylow=0.
    yhigh=max(Y3)+5

# Plotting data
fig=plt.figure()
ax=plt.subplot()
ax.tick_params(direction='in',which='both')
ax.plot(X,Y1,'m-',label='Mg')
ax.plot(X,Y2,'g-',label='O')
ax.plot(X,Y3,'k:',label='total')
ax.set(xlabel='$E-E_{fermi}$ (eV)', ylabel='$n(E-E_{fermi})$ (arb. unit)')
plt.axvline(x=0, color='r', linestyle='-')
plt.xlim(xlow,xhigh)
plt.ylim(ylow,yhigh)
ax.legend()
plt.show()
fig.savefig('MgO/MgO_projected_DOSS.png',dpi=600)
```
