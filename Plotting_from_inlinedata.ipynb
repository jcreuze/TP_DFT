```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = [0,1,2,3,4]
s1 = [4,3,2,1,0]
s2 = [2,1.5,1,0.5,0]

# Plotting data
fig=plt.figure()
bx=plt.subplot()
bx.set_yticks([0,1,2,3,4]) # Set the x ticks at positions [x1,x2,...,xn]
bx.set_xticks([0,1,2,3,4]) # Set the y ticks at positions [y1,y2,...,yn]
bx.tick_params(direction='in',which='both')
bx.plot(t,s1,'-',color=(1,0,0),label='label datalist 1') # Color in RGB format (Red,Green,Blue)
bx.plot(t,s2,'--',color=(0.1,0.4,0.7),label='label datalist 2') # Color in RGB format (Red,Green,Blue)
bx.set(xlabel='$physical\ quantity_{1}$ (units)', ylabel='$physical\ quantity^{2}$ (units)',
       title='If needed')
#bx.grid()
plt.xlim(0,4) # Set the x range
plt.ylim(0,4) # Set the y range
bx.legend()
plt.show()
#fig.savefig('example.png',dpi=600)
```
